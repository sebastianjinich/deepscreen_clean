```python
import pandas as pd
import requests

def get_protein_sequence(uniprot_accession):
    url = f"https://www.uniprot.org/uniprot/{uniprot_accession}.fasta"
    response = requests.get(url)
    response.raise_for_status()
    lines = response.text.split("\n")
    sequence = "".join(lines[1:])  # Concatenate all lines except the first (header)
    return sequence

def get_all_protein_sequences(uniprot_accessions):
    protein_sequences = []
    for accession in uniprot_accessions:
        try:
            sequence = get_protein_sequence(accession)
            protein_sequences.append({'Accession': accession, 'Sequence': sequence})
            print(f"Retrieved sequence for {accession}")
        except requests.exceptions.HTTPError as e:
            print(f"Error retrieving sequence for {accession}: {e}")
    return pd.DataFrame(protein_sequences)
```

```python

df = get_all_protein_sequences(accessions)
df.to_pickle('uniprot_mapping.pickle')
```
