```python
def compid_target_desduplication_median(df):
    
    group = ['comp_id', 'target', 'relation', 'type','organism','assay_type']
    median_bioactivity = df.groupby(group,dropna=False)['bioactivity'].median().reset_index()
    median_bioactivity = median_bioactivity.rename(columns={'bioactivity': 'median_bioactivity'})
    df = df.drop_duplicates(group)
    df = pd.merge(df, median_bioactivity, on=group, how='left')
    df['bioactivity'] = df['median_bioactivity']
    df = df.drop(columns=['median_bioactivity'])
    return df
```

```python
print('dp antes de median desduplication =',len(df))
df = compid_target_desduplication_median(df)
print('dp despues de median desduplication =',len(df))
```

```python
def binary_biactivity(df):
    def step_func_bioactivity(row):
        relation = row['relation']
        bioactivity = row['bioactivity']

        if relation in ('=','<') and bioactivity <= 10000:
            return 1

        elif relation in ('=','>') and bioactivity >= 20000:
            return 0

        else:
            return 'mal'
    
    df['binary_bioactivity'] = df.apply(step_func_bioactivity,axis=1)
    print('bbunique:',df.binary_bioactivity.unique())
    df = df[df.binary_bioactivity != 'mal']
    return df
```

```python
print('dp antes de binary bioactivity =',len(df))
df = binary_biactivity(df)
print('dp despues de binary bioactivity =',len(df))
```

```python
def compid_target_desduplication_binary_bioact(df):
    df.sort_values('type')

    groups = df.groupby(['comp_id', 'target'])

    # define a function to apply to each group
    def drop_duplicates(group):
        # get the unique values of binary_bioactivity in the group
        bioactivity_values = group['binary_bioactivity'].unique()

        # if there is only one unique value, keep the first row in the group
        if len(bioactivity_values) == 1:
            return group.head(1)

        # if there are multiple unique values, prioritize rows with Ki or Kd in the type column
        else:
            ki_or_kd_rows = group[group['type'].isin(['Ki', 'Kd'])]
            if len(ki_or_kd_rows) > 0:
                return ki_or_kd_rows.head(1)
            else:
                return pd.DataFrame()

    # apply the function to each group and concatenate the results
    df_clean = pd.concat([drop_duplicates(group) for _, group in groups])

    # reset the index of the DataFrame
    df_clean = df_clean.reset_index(drop=True)

    return df_clean

# def compid_target_desduplication_binary_bioact(df):
#     # Create an empty list to store cleaned DataFrames
#     cleaned_dfs = []
# 
#     # Iterate over each group of unique ('comp_id', 'target') pairs
#     for _, group in df.groupby(['comp_id', 'target']):
#         # Get the unique values of binary_bioactivity in the group
#         bioactivity_values = group['binary_bioactivity'].unique()
# 
#         # If there is only one unique value, keep the first row in the group
#         if len(bioactivity_values) == 1:
#             cleaned_dfs.append(group.head(1))
#         else:
#             # If there are multiple unique values, prioritize rows with Ki or Kd in the type column
#             ki_or_kd_rows = group[group['type'].isin(['Ki', 'Kd'])]
#             if len(ki_or_kd_rows) > 0:
#                 cleaned_dfs.append(ki_or_kd_rows.head(1))
# 
#     # Concatenate the cleaned DataFrames into a single DataFrame
#     df_clean = pd.concat(cleaned_dfs, ignore_index=True)
# 
#     return df_clean
```

```python
def balance_compuestos(df_input,carpeta_data_autores):
    df = df_input.copy()

    import os
    targets_autores_archivos = os.listdir(carpeta_data_autores)
    targets_autores = [i[:i.find('.pickle')] for i in targets_autores_archivos]
    
    balance_compuestos_positivos = pd.Series(dtype='int64')
    targets_ausentes = []

    for target in targets_autores:
        
        df_target = df[df.target == target]
        
        if len(df_target.index) > 1:
            mio_pos = df_target['comp_id'].unique()
            ensayos_mio = len(mio_pos)
        else:
            targets_ausentes.append(target)
            print(f'{target} ausente')
            continue
        
        try:
            df_autores = pd.read_pickle(os.path.join(carpeta_data_autores,f'{target}.pickle'))
            autores_pos = df_autores[df_autores[target] == 1]
            ensayos_autores = len(autores_pos.index)
            blance = ensayos_mio - ensayos_autores
            balance_compuestos_positivos[target]=blance
            print(f'{target} balance hecho')
        except:
            print(f'Unable to open {target}.pickle')
            continue
    
    return balance_compuestos_positivos, targets_ausentes
```
